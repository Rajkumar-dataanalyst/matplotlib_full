# MATPLOTLIB PROGRAMS TO UNDERSTAND 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data from your CSV or Excel
df = pd.read_excel(r'seaborn_dataset.xlsx')
```

```python
##################################################
# 1. Histogram (Distribution of Age)
##################################################

plt.figure()
plt.hist(df['age'], color='#ffcc5b', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
##################################################
# 2. Bar Plot (Average Salary by Department)
##################################################

avg_salary = df.groupby('department')['salary'].mean()

plt.figure()
avg_salary.plot(kind='bar', color='#ff7f50')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()


```

```python
##################################################
# 3. Line Plot (Experience vs Salary)
##################################################

sorted_df = df.sort_values('experience_years')

plt.figure()
plt.plot(sorted_df['experience_years'], sorted_df['salary'], marker='o', color='#ff69b4')
plt.title('Experience vs Salary')
plt.xlabel('Experience (yrs)')
plt.ylabel('Salary')
plt.show()


```

```python
##################################################
# 4. Scatter Plot (Weight vs Height)
##################################################

plt.figure()
plt.scatter(df['height'], df['weight'], color='#ff1493', marker='x')
plt.title('Weight vs Height')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
##################################################
# 5. Box Plot (Salary Distribution by Department)
##################################################

plt.figure()
df.boxplot(column='salary', by='department')
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.suptitle('')  # To clear the automatic title
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
##################################################
# 6. Violin Plot (Performance Score by Department) - Pure matplotlib doesn't have Violin directly
# So we can approximate with Box Plot instead or use "bxp"
##################################################

# We'll reuse Box Plot here
plt.figure()
df.boxplot(column='performance_score', by='department')
plt.title('Performance Score by Department')
plt.xlabel('Department')
plt.ylabel('Performance Score')
plt.suptitle('')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
##################################################
# 7. Horizontal Bar Plot (Count of Employees by Location)
##################################################

location_count = df['location'].value_counts()

plt.figure()
location_count.plot(kind='barh', color='#ff9f40')
plt.title('Count of Employees by Location')
plt.xlabel('Count')
plt.ylabel('Location')
plt.show()
```

```python


###############################################################
# 8. Pair-wise Plot (Using scatter for select pairs manually)
###############################################################

# Age vs Experience

plt.figure()
plt.scatter(df['age'], df['experience_years'], color='#ffccff')
plt.title('Age vs Experience')
plt.xlabel('Age')
plt.ylabel('Experience (yrs)')
plt.show()

# Age vs Performance Score

plt.figure()
plt.scatter(df['age'], df['performance_score'], color='#ff6666')
plt.title('Age vs Performance Score')
plt.xlabel('Age')
plt.ylabel('Performance Score')
plt.show()
```

```python


##################################################
# 9. Subplots (Multiple charts in 1)
##################################################
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

axs[0].hist(df['age'], color='#ffcc5b')
axs[0].set_title('Distribution of Age')

axs[1].bar(avg_salary.index, avg_salary, color='#ff7f50')
axs[1].set_title('Average Salary by Department')
axs[1].tick_params(rotation=45)

fig.tight_layout()
plt.show()
```

```python


##################################################
# 10. Annotations (Label points)
##################################################

plt.figure()
plt.plot(sorted_df['experience_years'], sorted_df['salary'], marker='o')
plt.title('Experience vs Salary')

# Annotate highest salary
max_idx = sorted_df['salary'].idxmax()
max_exp = sorted_df.loc[max_idx, 'experience_years']
max_sal = sorted_df.loc[max_idx, 'salary']

plt.annotate(f'max salaries:{max_sal}',
             (max_exp, max_sal),
             textcoords='offset points',
             xytext=(0,10),
             arrowprops=dict(arrowstyle='->'))
plt.xlabel('Experience (yrs)')
plt.ylabel('Salary')
plt.show()
```

# COMPLETE CUSTUMIZATION OF VISUALS

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data first
df = pd.read_excel(r'seaborn_dataset.xlsx')

# Prepare some aggregated or processed data first
avg_salary_by_dept = df.groupby('department')['salary'].mean()
location_count = df['location'].value_counts()

# Sort for better visualization
sorted_df = df.sort_values('experience_years')
```

```python
# ------------------------------------------------------------------------------
# 1. Line Plot (Customizations: color, marker, markersize, linestyle, linewidth, title, grid)

plt.figure()
plt.plot(sorted_df['experience_years'], sorted_df['salary'],
         color='#ff7f50', marker='o', markersize=7,
         linestyle='--', linewidth=2)
plt.title('Experience vs Salary (with custom color, marker, linestyle)')
plt.xlabel('Experience (yrs)')
plt.ylabel('Salary')
plt.grid(visible=True, color='#d3d3d3', linestyle='-', linewidth=0.5)
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 2. Scatter Plot (Customizations: color, marker, size, edgecolor, facecolor)

plt.figure()
plt.scatter(df['age'], df['performance_score'],
            color='#ffcc5b', marker='D', s=100,
            edgecolor='black')
plt.title('Performance Score vs Age')
plt.xlabel('Age')
plt.ylabel('Performance Score')
plt.grid(visible=False)
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 3. Bar Plot (Customizations: color, width, edgecolor)

plt.figure()
avg_salary_by_dept.plot(kind='bar', color='#ff9f40', width=0.7, edgecolor='black')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 4. Horizontal Bar Plot (Customizations: color, align, grid)

plt.figure()
location_count.plot(kind='barh', color='#ff6666', align='center')
plt.title('Count of Employees by Location')
plt.xlabel('Count')
plt.ylabel('Location')
plt.grid(visible=False)
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 5. Histogram (Customizations: color, edgecolor, density)

plt.figure()
plt.hist(df['age'], color='#ffccff', edgecolor='black', density=False, bins=10)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 6. Box Plot (Customizations: patch_artist, facecolor)

plt.figure()
box = plt.boxplot(df['salary'], patch_artist=True)
for patch in box['boxes']:
    patch.set_facecolor("#ff7f50")
    patch.set_edgecolor("black")
    patch.set_linewidth(2)

plt.title('Box Plot of Salary')
plt.ylabel('Salary')
plt.grid(False)
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 7. Pie Chart (Customizations: explode, shadow, startangle)

plt.figure()
location_count.plot(kind='pie',
                    explode=[0.1]*len(location_count),
                    shadow=True,
                    startangle=90,
                    autopct='%1.1f%%')
plt.title('Employee Distribution by Location')
plt.ylabel('')
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 8. Annotations (Customizations: arrow, text)

max_idx = sorted_df['salary'].idxmax()
max_exp = sorted_df.loc[max_idx, 'experience_years']
max_sal = sorted_df.loc[max_idx, 'salary']

plt.figure()
plt.plot(sorted_df['experience_years'], sorted_df['salary'], marker='o')
plt.annotate(f'max salaries:{max_sal}',
             (max_exp, max_sal),
             textcoords='offset points',
             xytext=(0,10),
             arrowprops=dict(arrowstyle='->', color='r'))
plt.title('Annotation Example')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 9. Subplots (Customizations: tight_layout, facecolor)

fig, axs = plt.subplots(2, 2, figsize=(14, 10), facecolor='#edf2f4')

axs[0, 0].hist(df['age'], color='#ffcc5b')
axs[0, 0].set_title('Distribution of Age')

axs[0, 1].bar(avg_salary_by_dept.index, avg_salary_by_dept, color='#ff9f40')
axs[0, 1].set_title('Average Salary by Department')
axs[0, 1].tick_params(rotation=45)

axs[1, 0].plot(sorted_df['experience_years'], sorted_df['salary'], color='#ff7f50', marker='o')
axs[1, 0].set_title('Experience vs Salary')

axs[1, 1].scatter(df['age'], df['performance_score'], color='#ff6666')
axs[1, 1].set_title('Performance Score vs Age')

fig.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------------------------
# 10. Legends (Customizations: title, location)

plt.figure()
plt.scatter(df['age'], df['salary'], color='#ffcc5b', label='age vs salary')
plt.legend(title='Legend Title', loc='lower right')
plt.title('Scatter Plot with Legend')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
